```python
import json
import requests
import pandas as pd
import csv
import time
```

```python
def fetch_data(loc):
    url = "https://zillow56.p.rapidapi.com/search"
    querystring = {"location": loc}

    headers = {
    "X-RapidAPI-Key": "0c737feaecmsh8f5d2858c8960adp1d4ca1jsnc7db9cb645d9",
    "X-RapidAPI-Host": "zillow56.p.rapidapi.com"
}
    
    res = requests.get(url, headers=headers, params=querystring)
    time.sleep(2) # wait for response
    print(res.text)
    temp = json.loads(res.text)
    if "results" in temp:
        return temp['results']
    else:
        return 'null'
    
```

```python
# df = pd.read_csv('uszips.csv')
file_path = 'bay_area_zipcode_list.txt'  # Replace with the actual path to your text file
zip_codes = []

with open(file_path, 'r') as file:
    for line in file.readlines():
        zip_codes.append(line.strip())
```

```python
all_responses = []
```

```python
for zip in zip_codes:
     data = fetch_data(zip)
     if data != 'null':
        all_responses.append(data)
```

```python
place = ['Houston', 'Miami', 'Tampa', 'Dallas', 'New York', 'Jersey City', 'Philadelphia']
```

```python

# for zip in list(df['zip']):
for loc in place:
    data_1 = fetch_data(loc)
    if data_1 != 'null':
        all_responses.append(data_1)
```

```python
property_list = []
for response in all_responses:
    results = response
    for result in results:
        property_list.append(result)
        
```

```python
df = pd.DataFrame(property_list)
```

```python
df.fillna("null", inplace=True)
```
